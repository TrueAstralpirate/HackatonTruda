```python
import pandas as pd
data = pd.read_csv("C:/Users/yaros/Downloads/vacs_train.csv", sep = ";")
```

```python
print(
    max(
        map(
            lambda x: len(x.split()),
            data['specializations'].to_numpy()
        )
    )
)
```

```python
d = dict({})
for i in data['specializations.names']:
    t = str(i)
    for j in t.split('|'):
        temp = t.split(' - ')
        first = temp[0]
        second = ' - '.join(temp[1::])
        if (first not in d.keys()):
            d[first] = [second]
        else:
            d[first].append(second)
print(len(d))
```

```python
for i in data['key_skills']:
    if str(i) != 'nan':
        print(i)
```

```python
# area_name(забендить, потом one-hot)
# city(забендить, потом one-hot)
# company_id(забендить, потом one-hot)
# salary_from(таргет)
# employment(one-hot)
# schedule(one-hot)
# experience(one-hot)
# key_skills
# specializations
data = data[['area.name', 'city', 'company.id', 'employment', 'schedule', 'experience', 'key_skills', 'specializations', 'salary_from']]
```

```python
data
```

```python
area_one_hot = pd.get_dummies(data['area.name'], prefix = 'area_name')
city_one_hot = pd.get_dummies(data['city'], prefix = 'city')
employment_one_hot = pd.get_dummies(data['employment'], prefix = 'employment')
schedule_one_hot = pd.get_dummies(data['schedule'], prefix = 'city')
experience_one_hot = pd.get_dummies(data['experience'], prefix = 'city')
```

```python
data_new = pd.concat([pd.get_dummies(data['area.name'], prefix = 'area_name'),\
                      pd.get_dummies(data['city'], prefix = 'city'),\
                     data['company.id'],\
                     pd.get_dummies(data['employment'], prefix = 'employment'),\
                     pd.get_dummies(data['schedule'], prefix = 'schedule'),\
                     pd.get_dummies(data['experience'], prefix = 'experience'),\
                      data['key_skills'],\
                      data['specializations'],\
                      data['salary_from']], axis=1, sort=False)
data_new
```

```python
s = set()
for spec in data['specializations']:
    for one in spec.split():
        s.add(one)
print(len(s))
print(s)
```

```python
data_new.to_csv("new_data.csv", index=False)
```
