```python
n, l = data.shape
res = []
count = 6
spec_names_dict = dict({})
for spec in data['specializations.names']:
    temp_res = []
    for req in str(spec).split('|'):
        requirement = req.strip()
        if requirement not in spec_names_dict.keys():
            spec_names_dict[requirement] = len(spec_names_dict) + 1
        temp_res.append(spec_names_dict[requirement])
    res.append(sorted(temp_res) + [0] * (count - len(temp_res)))
my_df = pd.DataFrame(res, columns=['field_post#' + str(i) for i in range(1, count + 1)])
for_dummies = list(spec_names_dict.keys())
dummies = pd.get_dummies(for_dummies)
dummies
```
